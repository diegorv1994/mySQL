```python
password=getpass.getpass('Introduce tu pass SQL... ')
engine=db.create_engine('mysql+pymysql://root:{}@localhost/publications'.format(password))
print ('Conectado al servidor SQL.')
```

```python
query='''
select group_concat(t.title) as titles, t.year
from 
(select title, year(max(pubdate)) as "year",
count(titleauthor.title_id) as numauthors,
group_concat(concat(authors.au_fname, " ", authors.au_lname)) as authors
from titles
left join titleauthor
on titles.title_id=titleauthor.title_id
left join authors
on authors.au_id=titleauthor.au_id
group by titles.title
order by numauthors desc) as t
group by t.year;
'''
```

```python
df=pd.read_sql_query(query, engine)
df
```

```python
import MySQLdb as mdb      # base de datos, conexion mysql


db_host='localhost'         #host
db_user='root'              #usuario
db_pass='password'          #password
db_name='Apps'              #nombre de la base de datos
con=mdb.connect(host=db_host, user=db_user, passwd=db_pass, db=db_name, autocommit=True, use_unicode=True, charset="utf8") #conecta con la base de datos
print ('Conectado.')
cur=con.cursor()            # cursor sql, para ejecutar comandos 


tabla ="""DROP TABLE IF EXISTS Ratings; 
          CREATE TABLE Ratings (
          id int,                        
          track_name varchar(255),    
          size_bytes bigint, 
          price float, 
          rating_count_tot float, 
          rating_count_ver float, 
          user_rating float, 
          user_rating_ver float, 
          prime_genre varchar(50))""" # se crea la tabla
          
          
cur.execute(tabla)
print ('Tabla creada.')
```

```python
from sqlalchemy import create_engine   # conexion mysql      
import pandas as pd                    # dataframe




datos=pd.read_csv('data/apple_store.csv')              # se leen los datos del archivo csv
print ('Datos leidos.')
#print (datos)

motor=create_engine('mysql+mysqlconnector://root:password@localhost:3306/Apps')   # motor de enlace

datos.to_sql(name='Ratings', con=motor, if_exists='append', index=False)           # datos a sql
print ('Datos copiados a SQL.')
```
